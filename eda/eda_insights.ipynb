# Data Loading and Type Casting
This section loads datasets and applies type casting using configurations defined in YAML files.

```python
# Load datasets
df_patients = pd.read_csv('../data/processed_data/processed_patients.csv')
df_encounters = pd.read_csv('../data/processed_data/processed_encounters.csv')
df_medications = pd.read_csv('../data/processed_data/processed_medications.csv')
df_procedures = pd.read_csv('../data/processed_data/processed_procedures.csv')
df_conditions = pd.read_csv('../data/processed_data/processed_conditions.csv')
```

```python
# ########
# Patients
# ########

# Load YAML column mappings 
filename_yaml = '../config/processed_patients.yaml' 
# Load dictionary mappings
dict_column_mappings_encounters = func_return_mappings(filename_yaml)
# Cast with YAML
df_patients = func_cast_columns(df_patients, dict_column_mappings_encounters['columns'])
```

```python
# ########
# Encounters
# ########
# Load YAML column mappings 
filename_yaml = '../config/processed_encounters.yaml' 
# Load dictionary mappings
dict_column_mappings_encounters = func_return_mappings(filename_yaml)
# Cast with YAML
df_encounters = func_cast_columns(df_encounters, dict_column_mappings_encounters['columns'])
```

```python
# ########
# Medications
# ########
# Load YAML column mappings 
filename_yaml = '../config/processed_medications.yaml' 
# Load dictionary mappings
dict_column_mappings_encounters = func_return_mappings(filename_yaml)
# Cast with YAML
df_medications = func_cast_columns(df_medications, dict_column_mappings_encounters['columns'])
```

```python
# ########
# Procedures
# ########
# Load YAML column mappings 
filename_yaml = '../config/processed_procedures.yaml' 
# Load dictionary mappings
dict_column_mappings_encounters = func_return_mappings(filename_yaml)
# Cast with YAML
df_procedures = func_cast_columns(df_procedures, dict_column_mappings_encounters['columns'])
```

```python
# ########
# Conditions
# ########
# Load YAML column mappings 
filename_yaml = '../config/processed_conditions.yaml' 
# Load dictionary mappings
dict_column_mappings_encounters = func_return_mappings(filename_yaml)
# Cast with YAML
df_conditions = func_cast_columns(df_conditions, dict_column_mappings_encounters['columns'])
```

# Distribution of patients by gender

```python
# Distribution of patients by gender
gender_counts = df_patients['gender'].value_counts()
plt.figure(figsize=(6, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Patients Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# Distribution of patients by RACE

*Finding*: Category "White" is dominating in the patients CSV

```python
# Distribution of patients by RACE
race_counts = df_patients['race'].value_counts()
plt.figure(figsize=(6, 6))
race_counts.plot(kind='bar')
plt.title('Race Distribution in Patients.csv')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()
```

# Distribution of patients by ethinicity

*Finding*: Category "nonhispanic" is dominating in the patients CSV

```python
# Distribution of patients by ethinicity
gender_counts = df_patients['ethnicity'].value_counts()
plt.figure(figsize=(6, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Patients Ethnicity Distribution')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.show()
```

# Distribution of patients by birthdate

*Finding*: Few patients are found who were born between 1989-1994 (an indication to missing data)

```python
# Distribution of patients by birthdate
df_patients['birth_year'] = df_patients['birthdate'].dt.year
# df_patients
patients_by_year = df_patients.groupby('birth_year').size()
# patients_by_year

# Creating bins of 5 years
bins = range(df_patients['birth_year'].min(), df_patients['birth_year'].max() + 5, 5) # goes from min to max in step of 5
df_patients['birth_year_bin'] = pd.cut(df_patients['birth_year'], bins=bins, right=False)
# df_patients.head(5)

# Grouping by year bins and counting the number of patients
patients_by_bin = df_patients.groupby('birth_year_bin').size()

# Plotting the histogram
plt.figure(figsize=(10, 6))
patients_by_bin.plot(kind='bar', width=0.8)
plt.title('Histogram of Patients by Birth Year (5-Year Bins)', fontsize=14)
plt.xlabel('Birth Year Bin', fontsize=12)
plt.ylabel('Number of Patients', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Trends of Patients Visits

*Finding*: Number of patient visiting facility hit the maximum around 2021-2022

```python
# Trends of Patients Visits
df_encounters['visit_year'] = df_encounters['start_time'].dt.year
visits_per_year = df_encounters.groupby('visit_year').size()
plt.figure(figsize=(10, 6))
visits_per_year.plot(kind='line', marker='o')
plt.title('Visits Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Visits')
plt.grid(True)
plt.show()
```

# Use of medicines over time

*Finding*: Trend of medications gradually increased from 2015

```python
# Use of medicines over time
df_medications['start_year'] = df_procedures['start_time'].dt.year
medications_per_year = df_medications.groupby('start_year').size()
# medications_per_year.head(12)
plt.figure(figsize=(10, 6))
medications_per_year.plot(kind='line', marker='o')
plt.title('Medications Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Medicines taken per year')
plt.grid(True)
plt.show()
```

# Trends of procedures over time

*Finding*: Trend of procedures (and medications) seems to pick up from 2015

```python
# Trends of procedures over time
df_procedures['procedure_year'] = df_procedures['start_time'].dt.year
procedures_per_year = df_procedures.groupby('procedure_year').size()
# procedures_per_year.head(12)

plt.figure(figsize=(10, 6))
procedures_per_year.plot(kind='line', marker='o')
plt.title('Procedures Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Procedures')
plt.grid(True)
plt.show()
```

# Most frequently occuring conditions

# Most frequent diagnosis

*Finding*: The majority of medications were prescribed for Hypertension, while the second most common diagnosis corresponds to missing data (labeled as Unknown).

```python
# Most frequent diagnosis
top_diagnosis = df_medications['reason_description_for_medication'].value_counts()
top_diagnosis.head(12)

plt.figure(figsize=(10, 6))
top_diagnosis.plot(kind='bar')
plt.title('Top 12 Diagnosis')
plt.xlabel('')
plt.ylabel('Frequency')

# Limit x-axis label characters to the first 100
labels = [str(label)[:20] for label in top_diagnosis.index]  # Convert labels to strings and truncate
plt.xticks(ticks=range(len(labels)), labels=labels, rotation=90)
plt.show()
```

# Top diagnosis by all genders

```python
# Top diagnosis by all genders
top_diagnosis_by_gender = df_medications.groupby(['reason_description_for_medication', 'gender']).size().unstack().fillna(0)
top_diagnosis_by_gender.head(5)
```

# Top diagnosis for Females

*Insight*: Females were predominantly prescribed medications for Hypertension, followed by Anemia.

```python
# Top diagnosis for Females
display_df = top_diagnosis_by_gender.sort_values(by=['F'], ascending=False)['F'].head(5)
display_df
```

# Top diagnosis for Males

*Insight*: Males were mainly prescribed medications for Hypertension, and childhood asthma appears to be the most common condition among them. However, data quality issues are present due to missing entries in the medication descriptions.

```python
# Top diagnosis for Males
display_df = top_diagnosis_by_gender.sort_values(by=['M'], ascending=False)['M'].head(5)
display_df
```
